BAR GRAPH

```python
year=[2008,2009,2010,2011,2012,2013,2014,2015]
sales_maruti=[1234323,4576891,7890563,4536753,8765490,6324378,7654870,8907609]
sales_honda=[2334523,9876450,1986784,9809987,7645321,1234583,9876452,6523128]
sales_mahindra=[9987652,1987763,6734521,1231234,6785423,9809879,1098765,6789301]
```

```python
plt.bar(year,sales_maruti,color='k',label='maruti',alpha=0.4)
plt.title("sales per year")
plt.xlabel("year")
plt.ylabel("units")
plt.grid()
plt.show()
```

```python
width=0.25
x_pos=np.arange(len(year))
plt.bar(x_pos-width,sales_maruti,color='r',label='maruti',width=width)
plt.bar(x_pos,sales_honda,color="b",label='honda',width=width)
plt.bar(x_pos+width,sales_mahindra,color='g',label='Mahindra',width=width)
plt.xticks(x_pos,year,rotation=60)
plt.legend(loc='best')
plt.show()
```

```python
a=[1,2,3,4,5]
b=[1,4,9,16,25]
```

```python
fig,ax=plt.subplots(2,2)
ax[0].plot(a,b)
ax[1].plot(a,b)
ax[2].plot(a,b)
ax[3].plot(a,b)
plt.tight_layout()
plt.show()
```

ANIMATION

```python
from matplotlib import animation
```

```python
fig=plt.figure()
ax=plt.subplot(1,1,1)
line1, =ax.plot([],[ ])

ax.set_xlim(0,2)
ax.set_ylim(-1.5,1.5)
x=np.linspace(0,2,1000)

def init():
  ax.set_xlabel("x")
  ax.set_ylabel("sinx")
  ax.set_title("sine curve")
  line1.set_data([],[])
  return line1,

def animate(i):
  y=np.sin(2*np.pi*(x-0.01*i))
  line1.set_data(x,y)
  return line1,

animation.FuncAnimation(fig,animate,init_func=init,frames=100,interval=100)
plt.show()
```
